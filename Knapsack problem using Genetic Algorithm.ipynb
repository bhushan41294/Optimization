```python
import numpy as np
import random
```

```python
num_items = int(input("Enter the number of items: "))
print("Please enter the weights with delimeter ',' for example 2,3,5,8,9")
weight = input("Enter the weights: ") 
print("Please enter the values with delimeter ',' for example 2,3,5,8,9")
value = input("Enter the values: ") 
max_weight = int(input("Enter the max. weight: "))
Total_generations=int(input("Enter number of iterations: ")) 
```

```python
w = [int(x) for x in weight.split(",")]
v = [int(x) for x in value.split(",")]
max_weight = max_weight
```

```python
population = np.random.randint(0,2,(len(w),len(v)))
Best_solution = [] 
for itr in range(Total_generations):
    for i in range(len(w)):
        sumWeight = sum(np.multiply(w,population[i,0:]))
        sumValue = sum(np.multiply(v,population[i,0:]))
        if sumWeight > max_weight:
            while True:
                population[i] = np.random.randint(0,2,len(w))
                sumWeight = sum(np.multiply(w,population[i,0:]))
                if sumWeight < max_weight:
                    break

    # For Cross over 
    r1 = np.random.randint(0,len(w))
    r2 = np.random.randint(0,len(w))

    if r1 == r2:
        while True:
            r2 = np.random.randint(0,4)
            if r1 != r2:
                break

    r1 = population[r1]
    r2 = population[r2]
    
    crossover_loc =  2
    
    crossover = np.concatenate((r1[0:crossover_loc], r2[crossover_loc:]),axis=0)
   

    #Mutation

    mutation_loc = np.random.randint(0,len(w))
    
    if crossover[mutation_loc] == 1:
        crossover[mutation_loc]=0
    else:
        crossover[mutation_loc]=1

    sumWeight = sum(np.multiply(w,crossover))
    sumValue = sum(np.multiply(v,crossover))
    if sumWeight <= max_weight:
        population[r1]=crossover
    else:
        population=population

    result = []
    for i in range(len(w)):
        sumWeight = sum(np.multiply(w,population[i,0:]))
        sumValue = sum(np.multiply(v,population[i,0:]))
        result.append({"Weight":sumWeight,"value":sumValue,"string":str(population[i])})

    z=[]
    for i in range(len(result)):
        z.append(sorted(result[i].items(),reverse=True))
    Result=sorted(z,reverse=True) 
    Best_solution.append(Result[0])
    print("The best solution for this generation is: ",Result[0])
    print("---------------------------------------------------------------------------------------------------")
    print(" ")
optimum_solution = sorted(Best_solution,reverse=True)[0]
print("The optimum solution is: ",optimum_solution)
```
